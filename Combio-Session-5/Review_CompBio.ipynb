```python
# read fasta
seqA = SeqIO.read('a.fasta', 'fasta').seq
seqB = SeqIO.read('b.fasta', 'fasta').seq

print(seqA)
print(seqB)
```

# 1. Sequence Manipulation

```python
#a. Find and display the total length of Sequence A and B. 

len_a = len(seqA)
len_b = len(seqB)
total_len = len_a + len_b
print(f"Total length sequence A and B = {total_len}")
```

```python
#b. Find and display which index the first time codon (TAC) appears in both Sequence A and B.

print(f'Index of first time codon "TAC" appears in Sequence A = {seqA.index('TAC')}')
print(f'Index of first time codon "TAC" appears in Sequence B = {seqB.index('TAC')}')
```

```python
#c. Take the first 10 nucleotides from Sequence A, and the last 15 nucleotides from Sequence B. Then combine the sequences and reverse it to a new sequence, Sequence C. 

#take the first 10 nucleotides from sequence A
ten_first = seqA[:10]

# the last 15 nucleotides from Sequence B
last_fifteen = seqB[-15:]

#combine
combine = ten_first + last_fifteen

#reverse
seqC = combine[::-1]

print(f'Sequence C = {seqC}')
```

# 2. Sequence Analysis & Plotting

```python
#a. Determine and display the nucleotides base (A, C, G, T) frequency for Sequence A, B, and C with a chart using matplotlib library. 

from collections import Counter
import matplotlib.pyplot as plt

counter_a = Counter(seqA)
counter_b = Counter(seqB)
counter_c = Counter(seqC)

plt.bar(counter_a.keys(), counter_a.values())
plt.title('GCAT Frequencies of Sequence A')
plt.show()

plt.bar(counter_b.keys(), counter_b.values())
plt.title('GCAT Frequencies of Sequence B')
plt.show()

plt.bar(counter_c.keys(), counter_c.values())
plt.title('GCAT Frequencies of Sequence C')
plt.show()
```

```python
#b. Determine and display the GC & AT content percentage for Sequence A, B, and C. 

from Bio.SeqUtils import gc_fraction

gc_a = gc_fraction(seqA)*100
print(f'GC Content of sequence A: {gc_a:.2f}%')
print(f'AT Content of sequence A: {100 -(gc_a):.2f}%')
print()

gc_b = gc_fraction(seqB)*100
print(f'GC Content of sequence B: {gc_b:.2f}%')
print(f'AT Content of sequence B: {100 - gc_b:.2f}%')
print()

gc_c = gc_fraction(seqC)*100
print(f'GC Content of sequence C: {gc_c:.2f}%')
print(f'AT Content of sequence C: {100 - gc_c:.2f}%')
print()
```

```python
#c. Determine and display the melting temperature for Sequence A, B, and C (using built-inWallace Method).

from Bio.SeqUtils import MeltingTemp as mt

mel_temp_a = mt.Tm_Wallace(seqA)
mel_temp_b = mt.Tm_Wallace(seqB)
mel_temp_c = mt.Tm_Wallace(seqC)

print(f'Melting Temperature of sequence A: {mel_temp_a}')
print(f'Melting Temperature of sequence B: {mel_temp_b}')
print(f'Melting Temperature of sequence C: {mel_temp_c}')
```

```python
#d. Determine and display the molecular weight for Sequence A, B, and C.

from Bio.SeqUtils import molecular_weight as mw

mol_weight_a = mw(seqA)
mol_weight_b = mw(seqB)
mol_weight_c = mw(seqC)

print(f'Molecular weight of sequence A: {mol_weight_a}')
print(f'Molecular weight of sequence B: {mol_weight_b}')
print(f'Molecular weight of sequence C: {mol_weight_c}')
```

# 3. DNA & mRNA Protein Synthesis

```python
#a. Transcribe and display the DNA Sequence A, B, and C into mRNA Sequences.
mRNA_a = seqA.transcribe()
mRNA_b = seqB.transcribe()
mRNA_c = seqC.transcribe()

print(f"mRNA A = {mRNA_a}")
print(f"mRNA B = {mRNA_b}")
print(f"mRNA C = {mRNA_c}")
```

```python
#b. Translate and display the mRNA Sequence A, B, and C into amino acids sequences. (Clue: for translate, use the mRNA sequences that already transcribed on the first step)

#nah ini sebenernya mau translate dari mRNA atau dari seqencenya langsung ga masalah, soalnya biopython masi bisa baca
amino_acids_a = mRNA_a.translate()
amino_acids_b = mRNA_b.translate()
amino_acids_c = mRNA_c.translate()

print(f'Amino Acid of sequence A: {amino_acids_a}')
print(f'Amino Acid of sequence B: {amino_acids_b}')
print(f'Amino Acid of sequence C: {amino_acids_c}')
```

# 4. Sequence Alignment & Similarities

```python
#a. Perform local alignment pairwise between Sequence A and Sequence “CGCTACTAGCATGCGCAT” and display the alignments’ score. (Clue: Use localxx) 

from Bio.pairwise2 import format_alignment, align

local_alignment_a = align.localxx(seqA,"CGCTACTAGCATGCGCAT")
for alignment in local_alignment_a:
    print(format_alignment(*alignment))
```

```python
#b. Perform global alignment pairwise between Sequence B and Sequence “CAGTGCGCTAGCAGCTACGTGTG” and display the alignments’ score. (Clue: Use globalxx)

from Bio import pairwise2
global_alignment_b = pairwise2.align.globalxx(seqB, 'CAGTGCGCTAGCAGCTACGTGTG')
for alignment in global_alignment_b:
    print(format_alignment(*alignment))
```

```python
#c. Find and display the Levenshtein distance between Sequence A and B
from Levenshtein import distance

distance(seqA, seqB)
```
