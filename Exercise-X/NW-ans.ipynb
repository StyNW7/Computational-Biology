# Import Library

```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
```

# Read Data

```python
seqA = SeqIO.read("Data/a.fasta", "fasta")
seqB = SeqIO.read("Data/b.fasta", "fasta")

seqA = seqA.seq
seqB = seqB.seq

print (f"Sequence A : {seqA}")
print (f"Sequence B : {seqB}")
```

# Sequence Manipulation

```python
print (f"Length of Sequence A : {len(seqA)}")
print (f"Length of Sequence B : {len(seqB)}")
```

```python
print (f"TAC First Time appears at Index: {seqA.find("TAC")} of Sequence A")
print (f"TAC First Time appears at Index: {seqB.find("TAC")} of Sequence B")
```

```python
seqA_spliced = seqA[:10]
seqB_spliced = seqB[-15:]

seqC = (seqA_spliced+seqB_spliced) [::-1]

print (f"Sequence C : {seqC}")
```

# Sequence Analysis and Plotting

```python
import matplotlib.pyplot as plt
from collections import Counter

counterSeqA = Counter(seqA)
counterSeqB = Counter(seqB)
counterSeqC = Counter(seqC)

plt.title("Nucleotides Frequency for Sequence A")
plt.bar(counterSeqA.keys(), counterSeqA.values())
plt.show()

plt.title("Nucleotides Frequency for Sequence B")
plt.bar(counterSeqB.keys(), counterSeqB.values())
plt.show()

plt.title("Nucleotides Frequency for Sequence C")
plt.bar(counterSeqC.keys(), counterSeqC.values())
plt.show()
```

```python
from Bio.SeqUtils import gc_fraction

gcSeqA = gc_fraction(seqA) * 100
gcSeqB = gc_fraction(seqB) * 100
gcSeqC = gc_fraction(seqC) * 100

atSeqA = 100 - gcSeqA
atSeqB = 100 - gcSeqB
atSeqC = 100 - gcSeqC

print (f"GC Percentage of Sequence A : {gcSeqA:.2f}%, AT Percentage of Sequence A : {atSeqA:.2f}%")
print (f"GC Percentage of Sequence B : {gcSeqB:.2f}%, AT Percentage of Sequence B : {atSeqB:.2f}%")
print (f"GC Percentage of Sequence C : {gcSeqC:.2f}%, AT Percentage of Sequence C : {atSeqC:.2f}%")
```

```python
from Bio.SeqUtils import MeltingTemp as mt

print (f"Melting Temperature of Sequence A : {mt.Tm_Wallace(seqA)}")
print (f"Melting Temperature of Sequence B : {mt.Tm_Wallace(seqB)}")
print (f"Melting Temperature of Sequence C : {mt.Tm_Wallace(seqC)}")
```

```python
from Bio.SeqUtils import molecular_weight

print (f"Molecular Weight of Sequence A : {molecular_weight(seqA):.2f}")
print (f"Molecular Weight of Sequence B : {molecular_weight(seqB):.2f}")
print (f"Molecular Weight of Sequence C : {molecular_weight(seqC):.2f}")
```

# DNA & mRNA Protein Synthesis

```python
mRNA_A = seqA.transcribe()
mRNA_B = seqB.transcribe()
mRNA_C = seqC.transcribe()

print (f"mRNA of Sequence A : {mRNA_A}")
print (f"mRNA of Sequence B : {mRNA_B}")
print (f"mRNA of Sequence C : {mRNA_C}")
```

```python
aminoAcid_A = mRNA_A.translate()
aminoAcid_B = mRNA_B.translate()
aminoAcid_C = mRNA_C.translate()

print (f"aminoAcid of Sequence A : {aminoAcid_A}")
print (f"aminoAcid of Sequence B : {aminoAcid_B}")
print (f"aminoAcid of Sequence C : {aminoAcid_C}")
```

# Sequence Alignment and Similarities

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
```

```python
alignment_seqA = pairwise2.align.localxx(seqA, "CGCTACTAGCATGCGCAT")

top_alignment = alignment_seqA[0]
print (format_alignment(*top_alignment))

print (f"Alignment Score of Sequence A : {top_alignment[2]}")
```

```python
alignment_seqB = pairwise2.align.globalxx(seqB, "CAGTGCGCTAGCAGCTACGTGTG")

top_alignment = alignment_seqB[0]
print (format_alignment(*top_alignment))

print (f"Alignment Score of Sequence B : {top_alignment[2]}")
```

```python
from Levenshtein import distance

print (f"Levenshtein Distance between Sequence A and Sequence B : {distance(seqA, seqB)}")
```
