#### Nucleotide Frequency Plot

```python
def nucleotide_frequency(seq):
    return {
        "A": seq.count("A"),
        "C": seq.count("C"),
        "G": seq.count("G"),
        "T": seq.count("T")
    }

freq = nucleotide_frequency(seq1)
print(freq)
```

```python
import matplotlib.pyplot as plt

def plot_frequency(seq, title="Nucleotide Frequency"):
    freq = nucleotide_frequency(seq)
    
    bases = ["A", "C", "G", "T"]
    values = [freq[b] for b in bases]
    
    plt.figure(figsize=(6,4))
    plt.bar(bases, values)
    plt.xlabel("Nucleotide")
    plt.ylabel("Count")
    plt.title(title)
    plt.show()

plot_frequency(seq1)
```
