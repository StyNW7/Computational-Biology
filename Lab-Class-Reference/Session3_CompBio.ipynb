```python
from Bio.Seq import Seq
```

```python
# DNA coding strand (sense strand)
dnaSeq = Seq('ATGATCTCGTAA')
dnaSeq
```

```python
# Complement adalah pasangan untai DNA berdasarkan aturan pasangan basa (A-T, C-G).
# DNA secara alami berbentuk double helix yang terdiri dari dua untai berpasangan.
#
# Dalam biologi molekuler, yang ditranskripsi adalah DNA template (antisense strand).
# Namun, Biopython mengasumsikan bahwa DNA yang diberikan adalah coding strand,
# sehingga fungsi transcribe() langsung menghasilkan mRNA 
# tanpa perlu menghitung template strand secara eksplisit.

# Jadi code di bawah ini cuma buat nunjukin pasangan dari sequence di atasnya,
# tapi nanti yang di transkripsi tetep sequence yang atas (coding strand)

# DNA template strand (antisense strand)
cDNASeq = dnaSeq.reverse_complement()
cDNASeq
```

```python
# Transkripsi = dna sense -> mRNA
# T -> U
mRnaSeq = dnaSeq.transcribe()
mRnaSeq
```

```python
# Translasi = mRNA -> protein (tapi hasilnya cuma huruf depan dari nama proteinnya aja)
# * = stop codon
aminoAcid = mRnaSeq.translate()
aminoAcid
```

```python
from Bio.SeqUtils import seq1, seq3
```

```python
# protein yang awalnya 1 huruf aja menjadi 3 huruf
polyPeptide = seq3(aminoAcid)
polyPeptide
```

```python
# balikin nama protein ke 1 huruf lagi
seq1(polyPeptide)
```

### Struktur 3D Protein

Struktur 3D protein dihasilkan dari proses **protein folding**, yaitu pelipatan rantai asam amino (polipeptida) di ruang tiga dimensi. Pada tahap awal setelah translasi, protein masih berupa rantai asam amino linear (polypeptide) yang belum memiliki fungsi biologis. Agar protein dapat menjalankan fungsinya, rantai ini harus melipat menjadi bentuk 3D yang spesifik dan stabil.

Proses pelipatan ini tidak terjadi secara acak. Bentuk akhir protein ditentukan oleh urutan asam amino dan interaksi fisika-kimia di antara asam amino tersebut. Ketika folding selesai, protein mencapai kondisi yang disebut native structure, yaitu struktur dengan energi paling rendah dan fungsi biologis yang optimal.

Salah satu faktor terpenting dalam protein folding adalah sifat hidrofobik dan hidrofilik dari asam amino:

* Asam amino hidrofobik (tidak suka air) cenderung menghindari lingkungan berair di dalam sel, sehingga akan terlipat ke bagian dalam (core) protein.

* Asam amino hidrofilik (suka air) lebih stabil jika berada di permukaan protein, sehingga dapat berinteraksi dengan air dan molekul lain di sekitarnya.

Distribusi asam amino hidrofobik dan hidrofilik inilah yang mendorong protein untuk melipat menjadi bentuk tertentu.
