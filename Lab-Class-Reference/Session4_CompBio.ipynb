```python
from Bio.Seq import Seq
from Bio import pairwise2
from Bio.pairwise2 import format_alignment as fa
from Levenshtein import distance
```

```python
seq1 = Seq("CGATGACAAGGCATGACGATACGATAATAGCATGCA")
seq2 = Seq("ATGACGAAATGACTAAGCAAAGCAATGCA")
```

##Local Alignment

Local aligntment = mencari subsequence terbaik, bukan seluruh sequence, dia dapet point dari match mismatch dan gap (bisa cari di youtube konsep lengkapnya)

ex:
#####seq1: ACGTACGT
#####seq2:     TACG

local alignment = TACG

```python
#localxx = scoring sederhana -> match +1, mismatch 0, gap 0
lA = pairwise2.align.localxx(seq1, seq2)
lA
"""
dia return dalam bentuk
[
  (aligned_seq1,
   aligned_seq2,
   score,
   begin,
   end)
]
"""
```

```python
# ini supaya lebih enak dibaca
for a in lA:
  # menggunakan format alignment'
  # * = unpacking operator, jadi intinya si fa itu minta 5 parameter, nah pake * supaya semua parameternya ke kirim
  print(fa(*a))
```

## Global Alignment
Global alignment menyelaraskan seluruh sequence dari awal sampai akhir

ex:

Sequence 1: ACGTACGT

Sequence 2: ACGCGT

Hasil global alignment :

ACGTACGT

ACG--CGT

```python
match_score = +2
mismatch_score = -1
gap_open = -0.5 # penalty saat gap pertama muncul
gap_extend = -0.1 # penalty tambahan jika gap panjang

#globalms = scoring custom
gAm = pairwise2.align.globalms(seq1, seq2,
                         match_score,
                         mismatch_score,
                         gap_open,
                         gap_extend)

for a in gAm:
    print(fa(*a))
```

```python
# global alignment dengan scoring sederhana -> match +1, mismatch 0, gap 0
gAx = pairwise2.align.globalxx(seq1,seq2)

for a in gAx:
    print(fa(*a))
```

##Hamming distance

Hamming distance adalah jumlah posisi yang berbeda antara dua sequence dengan panjang yang sama.

ex:
- s1 = ACGT
- s2 = AGGT

Simulasi
- A vs A → sama
- C vs G → beda
- G vs G → sama
- T vs T → sama

Hamming distance = 1

```python
def hd(s1, s2):
    list = []
    for x,y in zip(s1,s2):
        if x != y:
            list.append((x,y))

    return len(list)

hd(seq1,seq2)
```

```python
# versi lebih pendek
def hd2(s1,s2):
    return len([(x,y) for (x,y) in zip(s1, s2) if x != y])

hd2(seq1,seq2)
```

## Levenshtein distance

Levenshtein distance = jumlah minimum operasi berikut:

- Insertion → tambah karakter
- Deletion → hapus karakter
- Substitution → ganti karakter

untuk mengubah string A menjadi string B.

ex:
- seq1 = "ACGT"
- seq2 = "AGT"

perubahan:
- ACGT → AGT (hapus C)

distance = 1

```python
distance(seq1,seq2)
```

## Dot Plot
metode visual untuk membandingkan dua sequence untuk melihat tingkat kemiripan, perbedaan, dan pola tertentu.

```python
# buat matrix, kalo nucleotidanya sama kasih warna
def dotplot(s1,s2, nonblank = chr(0x25A0)):
    print(" |" + s2)
    print("-" * (len(s2) + 2))

    for x in s1:
        line = ''
        for y in s2:
            if x == y:
                line += nonblank
            else:
                line += ' '

        print(x + "|" + line)

dotplot(seq1,seq2)
```
