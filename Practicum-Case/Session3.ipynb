```python
from Bio.Seq import Seq

# =========================
# INPUT DNA SEQUENCES
# =========================
sequence_A = Seq("ATGCGTTTATGACGTATCTGAAGTCTACTAGGGTGATCTGACTAA")
sequence_B = Seq("GTGTTAGTGCGCTAGTGCTGACTAGGTACAGTAGCTAGTTACTAG")

# =========================
# 1️⃣ TRANSCRIBE DNA → mRNA
# =========================
mrna_A = sequence_A.transcribe()
mrna_B = sequence_B.transcribe()

print("mRNA Sequence A:", mrna_A)
print("mRNA Sequence B:", mrna_B)

# =========================
# 2️⃣ CREATE mRNA C
# (24 nt awal dari A + 21 nt akhir dari B)
# =========================
mrna_C = mrna_A[:24] + mrna_B[-21:]

print("\nmRNA Sequence C:", mrna_C)

# =========================
# 3️⃣ TRANSLATE mRNA → PROTEIN
# =========================
protein_A = mrna_A.translate()
protein_B = mrna_B.translate()
protein_C = mrna_C.translate()

print("\nProtein Sequence A:", protein_A)
print("Protein Sequence B:", protein_B)
print("Protein Sequence C:", protein_C)

# =========================
# 4️⃣ BACK TRANSCRIBE mRNA → DNA
# =========================
dna_C = mrna_C.back_transcribe()

print("\nBack Transcribed DNA C:", dna_C)

# =========================
# 5️⃣ COMPLEMENTARY DNA
# =========================
complement_A = sequence_A.complement()
complement_B = sequence_B.complement()
complement_C = dna_C.complement()

print("\nComplementary DNA A:", complement_A)
print("Complementary DNA B:", complement_B)
print("Complementary DNA C:", complement_C)
```
