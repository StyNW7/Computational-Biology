## Import Libary

```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
```

## Read Data

```python
seqA = SeqIO.read("Data/a.fasta", "fasta")
seqB = SeqIO.read("Data/b.fasta", "fasta")

# Change it to sequence.seq so that we can use seq method after this
seqA = seqA.seq
seqB = seqB.seq

print (seqA)
print (seqB)
```

## 1. Sequence Manipulation

a) Find and display the total length of Sequence A and B. 

```python
print (f"Length of Sequence A: {len(seqA)}")
print (f"Length of Sequence B: {len(seqB)}")
```

b) Find and display which index the first time codon (TAC) appears in both Sequence A and B. 

```python
print (f"TAC first time appears in sequence A at index: {seqA.find("TAC")}")
print (f"TAC first time appears in sequence B at index: {seqB.find("TAC")}")
```

c) Take the first 10 nucleotides from Sequence A, and the last 15 nucleotides from Sequence B. Then combine the sequences and reverse it to a new sequence, Sequence C.

```python
seqA_sliced = seqA[:10]
seqB_sliced = seqB[-15:]

seqC = (seqA_sliced + seqB_sliced) [::-1]

print (f"Sequence C: {seqC}")
```

## 2. Sequence Analysis and Plotting

a)  Determine and display the nucleotides base (A, C, G, T) frequency for Sequence A, B, and C with a chart using matplotlib library.

```python
import matplotlib.pyplot as plt
from collections import Counter

seqA_freq = Counter(seqA)
seqB_freq = Counter(seqB)
seqC_freq = Counter(seqC)

plt.title("Nucleotides base frequency for Sequence A")
plt.bar(seqA_freq.keys(), seqA_freq.values())
plt.show()

plt.title("Nucleotides base frequency for Sequence B")
plt.bar(seqB_freq.keys(), seqB_freq.values())
plt.show()

plt.title("Nucleotides base frequency for Sequence C")
plt.bar(seqC_freq.keys(), seqC_freq.values())
plt.show()
```

b) Determine and display the GC & AT content percentage for Sequence A, B, and C. 

```python
from Bio.SeqUtils import gc_fraction

seqA_GC = gc_fraction(seqA) * 100
seqB_GC = gc_fraction(seqB) * 100
seqC_GC = gc_fraction(seqC) * 100

seqA_AT = 100 - seqA_GC
seqB_AT = 100 - seqB_GC
seqC_AT = 100 - seqC_GC

print (f"Sequence A GC Percentage: {seqA_GC:.2f}%, Sequence A AT Percentage: {seqA_AT:.2f}%")
print (f"Sequence B GC Percentage: {seqB_GC:.2f}%, Sequence B AT Percentage: {seqB_AT:.2f}%")
print (f"Sequence C GC Percentage: {seqC_GC:.2f}%, Sequence C AT Percentage: {seqC_AT:.2f}%")
```

c) Determine and display the melting temperature for Sequence A, B, and C (using built-inWallace Method).  

```python
from Bio.SeqUtils import MeltingTemp as mt

print (f"Melting temperature for Sequence A: {mt.Tm_Wallace(seqA)}")
print (f"Melting temperature for Sequence B: {mt.Tm_Wallace(seqB)}")
print (f"Melting temperature for Sequence C: {mt.Tm_Wallace(seqC)}")
```

d) Determine and display the molecular weight for Sequence A, B, and C.

```python
from Bio.SeqUtils import molecular_weight

print (f"Molecular Weight for Sequence A: {molecular_weight(seqA):.2f}")
print (f"Molecular Weight for Sequence B: {molecular_weight(seqB):.2f}")
print (f"Molecular Weight for Sequence C: {molecular_weight(seqC):.2f}")
```

## 3. DNA and mRNA Protein Synthesis

a) Transcribe and display the DNA Sequence A, B, and C into mRNA Sequences. 

```python
mRNA_seqA = seqA.transcribe()
mRNA_seqB = seqB.transcribe()
mRNA_seqC = seqC.transcribe()

print (f"mRNA of Sequence A: {mRNA_seqA}")
print (f"mRNA of Sequence B: {mRNA_seqB}")
print (f"mRNA of Sequence C: {mRNA_seqC}")
```

b) Translate and display the mRNA Sequence A, B, and C into amino acids sequences. 
(Clue: for translate, use the mRNA sequences that already transcribed on the first 
step)

```python
amino_seqA = mRNA_seqA.translate()
amino_seqB = mRNA_seqB.translate()
amino_seqC = mRNA_seqC.translate()

print (f"Amino Acids of Sequence A: {amino_seqA}")
print (f"Amino Acids of Sequence B: {amino_seqB}")
print (f"Amino Acids of Sequence C: {amino_seqC}")
```

## 4. Sequence Alignment & Similarities

a) Perform local alignment pairwise between Sequence A and Sequence CGCTACTAGCATGCGCAT and display the alignments score. (Clue: Use localxx) 

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

alignment_seqA = pairwise2.align.localxx(seqA, "CGCTACTAGCATGCGCAT")

# Show Format
top_alignment = alignment_seqA[0]
print (format_alignment(*top_alignment))

# Show Score

print (f"Alignment Score of Seq A: {top_alignment[2]}")
```

b) Perform global alignment pairwise between Sequence B and Sequence CAGTGCGCTAGCAGCTACGTGTG and display the alignments score. (Clue: Use globalxx) 

```python
alignment_seqB = pairwise2.align.globalxx(seqB, "CAGTGCGCTAGCAGCTACGTGTG")

# Show Format
top_alignment = alignment_seqB[0]
print (format_alignment(*top_alignment))

print (f"Alignment Score of Sequence B: {top_alignment[2]}")
```

c. Find and display the Levenshtein distance between Sequence A and B

```python
from Levenshtein import distance

print (f"Levenshtein Distance between Sequence A and Sequence B: {distance(seqA, seqB)}")
```
