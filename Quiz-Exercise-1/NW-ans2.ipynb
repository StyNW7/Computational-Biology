# Import Library

```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
```

# Read Fasta

```python
seqA = SeqIO.read("Data/a.fasta", "fasta")
seqB = SeqIO.read("Data/b.fasta", "fasta")

seqA = seqA.seq
seqB = seqB.seq

print (f"Sequence A : {seqA}")
print (f"Sequence B : {seqB}")
```

# Sequence Manipulation

```python
print (f"Length of Sequence A : {len(seqA)}")
print (f"Length of Sequence B : {len(seqB)}")
```

```python
print (f"First Index Codon 'CGAT' in Sequence A : {seqA.find("CGAT")}")
print (f"First Index Codon 'CGAT' in Sequence B : {seqB.find("CGAT")}")
```

```python
seqA_sliced = seqA[:5]
seqB_sliced = seqB[-5:]

seqC = (seqA_sliced + seqB_sliced) [::-1]

print (f"Sequence C : {seqC}")
```

# Sequence Analysis and Plotting

```python
from Bio.SeqUtils import MeltingTemp as mt

print (f"Melting temperature of Sequence A : {mt.Tm_Wallace(seqA)}")
print (f"Melting temperature of Sequence B : {mt.Tm_Wallace(seqB)}")
print (f"Melting temperature of Sequence C : {mt.Tm_Wallace(seqC)}")
```

```python
from Bio.SeqUtils import gc_fraction

gc_seqA = gc_fraction(seqA) * 100
gc_seqB = gc_fraction(seqB) * 100
gc_seqC = gc_fraction(seqC) * 100

at_seqA = 100 - gc_seqA
at_seqB = 100 - gc_seqB
at_seqC = 100 - gc_seqC

print  (f"GC Percentage of Sequence A : {gc_seqA:.2f}%, AT Percentage of Sequence A : {at_seqA:.2f}%")
print  (f"GC Percentage of Sequence B : {gc_seqB:.2f}%, AT Percentage of Sequence B : {at_seqB:.2f}%")
print  (f"GC Percentage of Sequence C : {gc_seqC:.2f}%, AT Percentage of Sequence C : {at_seqC:.2f}%")
```

```python
from Bio.SeqUtils import molecular_weight

print (f"Molecular Weight of Sequence A : {molecular_weight(seqA):.2F}")
print (f"Molecular Weight of Sequence B : {molecular_weight(seqB):.2F}")
print (f"Molecular Weight of Sequence C : {molecular_weight(seqC):.2F}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

seqA_freq = Counter(seqA)
seqB_freq = Counter(seqB)
seqC_freq = Counter(seqC)

plt.title("Nucleotides Base of Sequence A")
plt.bar(seqA_freq.keys(), seqA_freq.values())
plt.show()

plt.title("Nucleotides Base of Sequence B")
plt.bar(seqB_freq.keys(), seqB_freq.values())
plt.show()

plt.title("Nucleotides Base of Sequence C")
plt.bar(seqC_freq.keys(), seqC_freq.values())
plt.show()
```

# DNA and mRNA Protein Synthesis

```python
mRNA_seqA = seqA.transcribe()
mRNA_seqB = seqB.transcribe()
mRNA_seqC = seqC.transcribe()

print (f"mRNA of Sequence A : {mRNA_seqA}")
print (f"mRNA of Sequence B : {mRNA_seqB}")
print (f"mRNA of Sequence C : {mRNA_seqC}")
```

```python
protein_seqA = mRNA_seqA.translate()
protein_seqB = mRNA_seqB.translate()
protein_seqC = mRNA_seqC.translate()

print (f"Protein of Sequence A : {protein_seqA}")
print (f"Protein of Sequence B : {protein_seqB}")
print (f"Protein of Sequence C : {protein_seqC}")
```

```python
from Bio.SeqUtils import seq3

protein_seqA = mRNA_seqA.translate()
protein_seqB = mRNA_seqB.translate()
protein_seqC = mRNA_seqC.translate()

print (f"Protein of Sequence A : {seq3(protein_seqA)}")
print (f"Protein of Sequence B : {seq3(protein_seqB)}")
print (f"Protein of Sequence C : {seq3(protein_seqC)}")
```

# Sequence Alignment and Similarities

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
```

```python
alignment_seqA = pairwise2.align.localxx(seqA, seqC)

top_alignment = alignment_seqA[0]

print (f"Sequence A Alignment : {format_alignment(*top_alignment)}")

print (f"Sequence A Score: {top_alignment[2]}")
```

```python
alignment_seqB = pairwise2.align.globalxx(seqB, seqC)

top_alignment = alignment_seqB[0]

print (f"Sequence B Alignment : {format_alignment(*top_alignment)}")

print (f"Sequence B Score: {top_alignment[2]}")
```

```python
from Levenshtein import distance

print (f"Levensthein Distance between Sequence A and Sequence B : {distance(seqA, seqB)}")
```
