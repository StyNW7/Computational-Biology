# Import Library

```python
import Bio
from Bio.Seq import Seq
from Bio import SeqIO
```

# Read Data

```python
seqA = SeqIO.read("Data/a.fasta", "fasta")
seqB = SeqIO.read("Data/b.fasta", "fasta")

seqA = seqA.seq
seqB = seqB.seq

print (f"Sequence A : {seqA}")
print (f"Sequence B : {seqB}")
```

# Sequence Combination

```python
seqA_sliced = seqA[:15]
seqB_sliced = seqB[-30:]

seqC = (seqA_sliced + seqB_sliced) [::-1]

print (f"Sequence C : {seqC}")
```

# String Manipulation

```python
print(f"Length of Sequence A : {len(seqA)}")
print(f"Reversed Sequence C : {seqC[::-1]}")
print(f"First index of T in Sequence B : {seqB.find('T')}")
print(f"Total number of A in Sequence A : {seqA.count('A')}")
print(f"Sequence C from index 5-10 : {seqC[5:11]}")
```

# Melting Temperature

```python
from Bio.SeqUtils import MeltingTemp as mt

MeltingTempA = mt.Tm_Wallace(seqA)
MeltingTempB = mt.Tm_Wallace(seqB)
MeltingTempC = mt.Tm_Wallace(seqC)

print (f"Melting Temperature of Sequence A : {MeltingTempA}")
print (f"Melting Temperature of Sequence B : {MeltingTempB}")
print (f"Melting Temperature of Sequence C : {MeltingTempC}")
```

# Molecular Weight

```python
from Bio.SeqUtils import molecular_weight

MolecularWeightA = molecular_weight(seqA)
MolecularWeightB = molecular_weight(seqB)
MolecularWeightC = molecular_weight(seqC)

print (f"Molecular Weight of Sequence A : {MolecularWeightA:.2f}")
print (f"Molecular Weight of Sequence B : {MolecularWeightB:.2f}")
print (f"Molecular Weight of Sequence C : {MolecularWeightC:.2f}")
```

# GC & AT Content Percentage

```python
from Bio.SeqUtils import gc_fraction

GCseqA = gc_fraction(seqA)
GCseqB = gc_fraction(seqB)
GCseqC = gc_fraction(seqC)

ATSeqA = 100 - GCseqA
ATSeqB = 100 - GCseqB
ATSeqC = 100 - GCseqC

print (f"GC Content Percentage of Sequence A : {GCseqA}%, AT Content Percentage of Sequence A : {ATSeqA}%")
print (f"GC Content Percentage of Sequence B : {GCseqB}%, AT Content Percentage of Sequence B : {ATSeqB}%")
print (f"GC Content Percentage of Sequence C : {GCseqC}%, AT Content Percentage of Sequence C : {ATSeqC}%")
```

# Nucleotides Base Frequency

```python
import matplotlib.pyplot as plt
from collections import Counter

seqA_freq = Counter(seqA)
seqB_freq = Counter(seqB)
seqC_freq = Counter(seqC)

print (f"Sequence A Frequency : {seqA_freq}")
print (f"Sequence B Frequency : {seqB_freq}")
print (f"Sequence C Frequency : {seqC_freq}")
```

# Plotting

## Molecular Weight

```python
plt.title("Molecular Weight of Sequence A and B")
plt.bar(["Sequence A", "Sequence B"], [MolecularWeightA, MolecularWeightB], color=["red", "blue"])
plt.show()
```

## GCAT Content Percentage

```python
plt.title("GC and AT Content Percentage of Sequence A")
plt.bar(["GC", "AT"], [GCseqA, ATSeqA])
plt.show()

plt.title("GC and AT Content Percentage of Sequence B")
plt.bar(["GC", "AT"], [GCseqB, ATSeqB])
plt.show()
```

## Nucleotides Base

```python
plt.title("GC and AT Content Percentage of Sequence A")
plt.bar(seqA_freq.keys(), seqA_freq.values())
plt.show()

plt.title("GC and AT Content Percentage of Sequence B")
plt.bar(seqB_freq.keys(), seqB_freq.values())
plt.show()
```

# Transcribtion and Translation

```python
mRNA_seqA = seqA.transcribe()
mRNA_seqB = seqB.transcribe()
mRNA_seqC = seqC.transcribe()

print (f"mRNA of Sequence A : {mRNA_seqA}")
print (f"mRNA of Sequence B : {mRNA_seqB}")
print (f"mRNA of Sequence C : {mRNA_seqC}")
```

```python
amino_acid_seqA = mRNA_seqA.translate(stop_symbol="#")
amino_acid_seqB = mRNA_seqB.translate(stop_symbol="#")
amino_acid_seqC = mRNA_seqC.translate(stop_symbol="#")

print (f"amino_acid of Sequence A : {amino_acid_seqA}")
print (f"amino_acid of Sequence B : {amino_acid_seqB}")
print (f"amino_acid of Sequence C : {amino_acid_seqC}")
```

```python
from Bio.SeqUtils import seq3
print(f"Amino acid translation of Sequence A : {seq3(mRNA_seqA.translate())}")
print(f"Amino acid translation of Sequence B : {seq3(mRNA_seqB.translate())}")
print(f"Amino acid translation of Sequence C : {seq3(mRNA_seqC.translate())}")
```

# Levenshtein Distance

```python
from Levenshtein import distance

seqD = Seq("ATCGATGCATGACGCATCGATGGCGATCGATTACGTGACTTACTG")

print (f"Levenshtein Distance between Sequence A and Sequence D : {distance(seqA[:45], seqD)}")
print (f"Levenshtein Distance between Sequence B and Sequence D : {distance(seqB[:45], seqD)}")
```

# Local and Global Alignment

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment 
```

```python
localAlignmentA = pairwise2.align.localxx(seqD, seqA)
topAlignment = localAlignmentA[0]

print (f"Local Alignment between Sequence A and D : {format_alignment(*topAlignment)}")
```

```python
localAlignmentB = pairwise2.align.localxx(seqD, seqB)
topAlignment = localAlignmentB[0]

print (f"Local Alignment between Sequence B and D : {format_alignment(*topAlignment)}")
```
